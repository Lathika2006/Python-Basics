# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('Sin Wave')
plt.show()
```

```python
x=np.linspace(-20,20,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r*')
plt.show()
```

```python
plt.plot(x,y,'b.')
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
x=np.linspace(-4,4,40)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.show()
```

Bar chat Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=['Tamil','English','Malayalam','Telungu','Kanada']
y=[75,120,40,45,70]
ax.bar(x,y)
plt.xlabel('Languages')
plt.ylabel('Speaking')
plt.title('Speaking Ratio')
plt.show()
```

Pie chart Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=['Tamil','English','Malayalam','Telungu','Kanada']
y=[75,120,40,45,70]
ax.pie(y,labels=x,autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,20,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='b')
plt.show()
```

3D plot

```python
ax=plt.axes(projection='3d')
x=20*np.random.random(100)
y=np.sin(x)
z=np.cos(x)
ax.scatter(y,z,x,c=x,cmap='Blues')
plt.show()
```
