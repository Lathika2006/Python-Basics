Manipulating a Dataframe

```python
cal_df.head()
```

```python
cal_df['Price']=california.target
```

```python
# Removing row(temporary removing)(axis=0 represents a row ,axis=1 represents column)
cal_df.drop(index=0,axis=0)
```

```python
# locating a particular column
print(cal_df.iloc[:,0]) # first column
print(cal_df.iloc[:,1]) # Second column
print(cal_df.iloc[:,-1]) # Last column
```
